```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Assuming matches is your existing DataFrame

# Define the columns to be scaled
strength_columns = ['attack_strength', 'defense_strength']

# Create a copy of the strength columns
strength_data = matches[strength_columns].copy()

# Initialize the scaler
scaler = StandardScaler()

# Fit and transform the strength data
scaled_strength_data = scaler.fit_transform(strength_data)

# Create a new DataFrame with the scaled strength values
scaled_strength_df = pd.DataFrame(scaled_strength_data, columns=strength_columns, index=matches.index)

# Replace the original strength columns with the scaled values in the original DataFrame
matches[strength_columns] = scaled_strength_df

# Now matches contains the scaled attack_strength and defense_strength
print(matches.head())  # Display the first few rows to verify
```

```python
grouped_matches = matches.groupby("team")
group = grouped_matches.get_group("Manchester City").sort_values("date")
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(5, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group

#cols = ["sh_x","xg_y","gf_x","sh_y","sot","att","tkl%","tklw","int","clr","tkl","err","lost","ga_x","gls","fk", "pk", "pkatt"]
cols = ["fk","gls","sot","sh_y","xg_y","pkatt","pass","err","tkl","ga_x","clr","int","blocks","tklw"]
#cols = ["xg_y","gf_x","sh_y","sot","err","lost","ga_x","gls","pass","blocks","dist","fk", "pk", "pkatt"]
new_cols = [f"{c}_rolling" for c in cols]

rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
```
